```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
for page in range(1,10):
    url = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=laptops&_sacat=0&_pgn=" + str(page)
    r = requests.get(url)
    print(r)


Soup = BeautifulSoup(r.content,"lxml")

print(Soup.prettify())
```

```python
NAMES = []
PRICES=[]
for page in range(1,10):
    url = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=laptops&_sacat=0&_pgn=" + str(page)
    r = requests.get(url)
    Soup = BeautifulSoup(r.content,"lxml")
    names=Soup.find_all("div", class_="s-item__title")

    for name in names:
        NAMES.append(name.text)

    prices=Soup.find_all("span",class_="s-item__price")
    for price in prices:
        PRICES.append(price.text)




#print(NAMES)
print(len(NAMES))
print(len(PRICES))
```

```python
laptops = pd.DataFrame({"Name":NAMES,"Price":PRICES})
print(laptops)
```

```python
csv_file_path = 'laptops.csv'
laptops.to_csv(csv_file_path, index=False)

print(f'DataFrame saved to {csv_file_path}')
```
